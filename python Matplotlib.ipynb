```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
x
```

```python
y = 3*x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=5)
plt.title('x vs y')
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.show()
```

```python
x
```

```python
y1 = 2*x
```

```python
y2 = 3*x
```

```python
plt.plot(x,y1,color='r',linestyle=':')
plt.plot(x,y2,color='y',linewidth=4)
plt.title('Two lines in a plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='r',linestyle=':')

plt.subplot(2,1,2)
plt.plot(x,y2,color='y',linewidth=4)
plt.title('Two line in a plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

```python
student = {'Biswajt':87,'Papu':57,'Guddy':100,'Babita':85}
```

```python
names=list(student.keys())
```

```python
names
```

```python
values = list(student.values())
```

```python
values
```

```python
plt.bar(names,values)
plt.show()
```

```python
plt.bar(names,values)
plt.title('Distribution of marks')
plt.xlabel('Names of students')
plt.ylabel('Marks of student')
plt.show()
```

```python
plt.barh(names,values,color="g")
plt.title('Distribution of marks')
plt.xlabel('Names of student')
plt.ylabel('Marks of student')
plt.show()
```

```python
fruit = ['Apple','Mango','Guava','Orange','Pineapple']
```

```python
quantity = [10,15,20,30,50]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','Red','Green','orange'])
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([100],radius=1,colors=['w'])
plt.show()
```
